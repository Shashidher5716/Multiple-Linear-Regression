```python
# Standarization using scaler min = 0, max = 1

def norm_fun(i):
  x = (i - i.min())/(i.max() - i.min())

  return x
```

```python
data_norm = norm_fun(data)
data = pd.DataFrame(data_norm)
data.describe()
```

```python
# Checking the correlation 

data.corr()
```

```python
# Changing the Col names

data.columns = ['Un', 'Pr', 'Sp', 'Hd','Ra', 'Sc']
```

```python
data
```

```python
# Using the Multi Linear Regression

model = smf.ols('Pr ~ Un + Sp + Hd + Ra + Sc', data = data).fit()
model.summary()
```

```python
pred1 = model.predict(data)
pred1
```

```python
final_model = smf.ols('Pr ~ Un + Sp + Hd + Ra + Sc', data = data).fit()
final_model.summary()
```

```python
# Predictions

pred = final_model.predict(data)
```

```python
# Q Q Plot
res = final_model.resid
sm.qqplot(res)
plt.show()
```

```python
# Residuals vs Fitted plot
sns.residplot(x = pred, y = data.Pr, lowess = True)
plt.xlabel('Fitted')
plt.ylabel('Residual')
plt.title('Fitted vs Residual')
plt.show()

sm.graphics.influence_plot(final_model)
```

```python
### Splitting the data into train and test data 
from sklearn.model_selection import train_test_split
data_train, data_test = train_test_split(data, test_size = 0.2) # 20% test data

# preparing the model on train data 
model_train = smf.ols('Pr ~ Un + Sp + Hd + Ra + Sc', data = data_train).fit()
model_train.summary()
```

```python
# prediction on test data set 
test_pred = model_train.predict(data_test)

# test residual values 
test_resid = test_pred - data_test.Pr
# RMSE value for test data 
test_rmse = np.sqrt(np.mean(test_resid * test_resid))
test_rmse #0.07
```

```python
# prediction on test data set 
train_pred = model_train.predict(data_train)

# test residual values 
train_resid = train_pred - data_train.Pr
# RMSE value for test data 
train_rmse = np.sqrt(np.mean(train_resid * train_resid))
train_rmse #0.07
```

```python
# Result : p_values for less than 0.05 &  R^2 =0.8
#           train_rmse is 0.07 & test_rmse is 0.07

#           Concluding the data is right fitting
```
